```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
fig=plt.figure(num=1, figsize=(12, 16))

ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
iterations=100
for i in range(iterations):
    j=int(-(i/10 -10)**2 +100)
    signal=np.random.uniform(i-0.1, i+0.1, 1000)
    phase=np.arange(0, 100, 0.1)
    signal[j*10:(j+1)*10]=signal[j*10:(j+1)*10]+0.5
    ax1.plot(phase, signal, color='k')
    #print(i,j)

base_signal=np.random.uniform(0.5, 0.75, 1000)
base_phase=np.arange(0, 1, 0.001)
base_signal[50:150]=base_signal[50:150]+4
ax2.plot(base_phase, base_signal)
ax2.set_ylim(0, 10)
plt.tight_layout()
```
